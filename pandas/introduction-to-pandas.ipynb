### .loc & .iloc

```python
# .loc & .iloc
animals = pd.Series(["cat", "dog", "bird", "panda", "snake"],
                    index=[0, 3, 9, 8, 3])
```

```python
animals
```

```python
#.loc is index label-based
animals.loc[3] 
```

```python
animals.loc[9]
```

```python
# .iloc refers to index or integer position based 
animals.iloc[3]
```

```python
car_sales
```

```python
# On dataframe
car_sales.loc[3]
```

```python
car_sales.iloc[3]
```

Here index and postions are both same, that's why might be confusing. 
Let's take another example

```python
# Dataframes with labeled index
car_data = car_sales.set_index('Make')
car_data
```

```python
# .loc refers to index label
car_data.loc['Nissan'] # Returns all records of index "Nissan".
```

```python
# .loc refers to index label
car_data.loc[3] # Gives error
```

```python
# .iloc refers to index position
car_data.iloc[3] # Returns all records at index 3
```
