# Changing Column names

```python
car_sales.rename(columns = {'Colour':'Color', 'Odometer (KM)': 'OdometerKM'}) # Changes won't apply until 'inplace = True'
```

## Adding Column(s)

```python
car_sales['Model'] = pd.Series(['Camary', 'Accord', 'MarkX', 'x7', 'Maxima', 'Grande', 'Civic', 'Civic'])
car_sales
```

```python
car_sales['YearRelease'] = pd.Series([1999, 2004, 2005, 2018, 2006, 2010, 2002, 1996, 2007, 2012])
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# Adding column using  calculation
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

## Adding Row(s)

```python
car_sales.append({'Make': 'Mercedes', 'Colour': 'Black', 'Odometer (KM)':'28092', 'Doors': 4}, ignore_index=True)
```

```python
new_cars = {
    'Make' : ['Audi', 'Hyundai', 'Lexus'],
    'Colour' : ['Black', 'Blue', 'White'],
    'Odometer (KM)' : [19200, 39000, 109900],
    'Price' : ['$12000', '$9500', '$22000']
}
car2 = pd.DataFrame(new_cars)
car2
```

```python
car_sales.append(car2, ignore_index = True)
```

```python
pd.concat([car_sales, car2], sort=False)
```
